```python
# Step 1: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

#Step 2: Load Dataset
df = pd.read_csv('/content/matches - matches.csv')
df.head()
```

###>> Know what kind of data we’re dealing with



Identify missing values

Get sum of total missing values column wise

Drop column with NULL values


```python
print(df.columns.tolist())
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
#Drop Umpire Column as whole column is empty

df.drop(columns=['umpire3'], inplace=True)
```

```python

# Fill missing values in 'city' with the most common city (mode)
df['city'] = df['city'].fillna('Unknown')

#Fill Missing Values in umpire1 and umpire2 column
df['umpire1'] = df['umpire1'].ffill()
df['umpire2'] = df['umpire2'].ffill()

df
```

## 📊 Dataset Overview
We first understand the structure and content of the dataset.

```python
df.shape
df.columns
df.info()
df.describe()
```

#City-wise Match Distribution
Which cities hosted the most IPL matches?


```python
df['city'].value_counts().plot(kind='barh', figsize=(10, 6), title='Number of Matches per City', color='teal')
plt.xlabel("Matches")
plt.ylabel("City")
plt.grid(True)
plt.show()
```

## 🏆 Team Performance (Wins)
Which teams won the most IPL matches overall?


```python
df['winner'].value_counts().plot(kind='bar', figsize=(10,6), title='Total Matches Won by Teams', color='orange')
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Insight
Mumbai Indians is the most successful IPL team in this dataset.

```python
df['winner'].value_counts().plot(kind='barh', figsize=(8,6), color='mediumpurple')
plt.title("Total Wins by Each Team")
plt.xlabel("Wins")
plt.ylabel("Team")
plt.show()
```

## Toss Decisions
Let’s examine toss decisions — how many times did teams choose to bat or field after winning the toss?


```python
df['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Toss Decision Distribution', colors=['lightblue', 'lightgreen'])
plt.ylabel("")
plt.show()
```

## Toss Winner vs Match Winner
How often did the toss-winning team actually win the match?


```python
(df['toss_winner'] == df['winner']).value_counts().plot(kind='bar', color='violet', title='Toss Winner = Match Winner?')
plt.xticks([0,1], ['No', 'Yes'], rotation=0)
plt.ylabel("Matches")
plt.grid(True)
plt.show()
```

## 📅 Season-wise Match Distribution
Which IPL seasons had the most matches?

```python
df['season'].value_counts().sort_index().plot(kind='line', marker='o', figsize=(10,5), title='Matches per IPL Season', color='crimson')
plt.xlabel("Season")
plt.ylabel("Matches")
plt.grid(True)
plt.show()
```

## Match Results and D/L Method Usage
Explore how matches ended and how often the Duckworth–Lewis (D/L) method was applied.


```python
# Result types (normal, tie, no result)
df['result'].value_counts().plot(kind='bar', color='skyblue', title='Match Result Types')
plt.ylabel("Matches")
plt.xticks(rotation=0)
plt.grid(True)
plt.show()

# D/L Method Usage
dl_matches = df[df['dl_applied'] == 1]
print("Total matches with D/L method applied:", dl_matches.shape[0])
```

## Final Insights
- **Mumbai Indians** won the most matches.
- Most matches were hosted in **Mumbai**, **Kolkata**, and   **Bangalore**.
- Toss winners won the match around **50%** of the time.
- Teams preferred to **field** after winning the toss.
- D/L method was rarely applied (less than 10 matches).
- 2013 and 2015 had the highest number of matches.
